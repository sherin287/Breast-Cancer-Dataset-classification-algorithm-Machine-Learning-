### Feature Scaling

## Classification Algorithm Implementation 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Logistic Regression
-  <font size="4">Logistic Regression is a linear model used for binary classification. It calculates the probability that a given input belongs to a particular class using the logistic (sigmoid) function. Logistic regression assumes a linear relationship between the features and the log-odds of the target.</font>
-  <font size="4">Efficient for datasets with many features, like the breast cancer dataset.</font>


### Decision Tree Classifier
- <font size="4"> Decision Trees partition the data into subsets based on feature values, forming a tree structure.</font>
-  <font size="4">Handles both numerical and categorical features.</font>
-  <font size="4">Captures non-linear relationships in the data.</font>

### Random Forest Classifier
-  <font size="4">Random Forest is an ensemble method that creates multiple decision trees during training and combines their predictions to improve accuracy and reduce overfitting.</font>
-  <font size="4">Effective for datasets with many features, such as the breast cancer dataset.</font>

#### Support Vector Machine (SVM)
-  <font size="4">SVM finds a hyperplane in a high-dimensional space that maximally separates the classes.It uses a kernel trick to handle non-linear decision boundaries.</font>

### k-Nearest Neighbors (k-NN)
-  <font size="4">k-NN is a distance-based algorithm where a sample is classified based on the majority class of its k-nearest neighbors in the feature space.Effective for smaller datasets where feature scaling has been applied.</font>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predictions

y_pred = model.predict(X_test)
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Logistic Regression Model Accuracy: {accuracy:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

### Implementing other models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest Classifier': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Classifier': SVC(),
    'K-Nearest Neighbors Classifier': KNeighborsClassifier(),
    'Decision Tree Classifier': DecisionTreeClassifier()
}
# Train and evaluate models
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, output_dict=True)

    # Extract metrics from classification report
    metrics = {
        'Model': name,
        'Accuracy': accuracy,
        'Precision': class_report['weighted avg']['precision'],
        'Recall': class_report['weighted avg']['recall'],
        'F1-Score': class_report['weighted avg']['f1-score']
    }
    results.append(metrics)

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Sort the DataFrame by accuracy in descending order
results_df_sorted = results_df.sort_values(by='Accuracy', ascending=False)

# Print the sorted DataFrame
(results_df_sorted)
```

### Model Comparison 

### Best Performers: Logistic Regression and Support Vector Classifier

-   <font size="5">Both models achieved the highest accuracy (97.37%) and identical scores for precision, recall, and F1-Score.
These algorithms are effective in handling linearly separable data, which likely contributed to their success on the breast cancer dataset.</font>
-  <font size="5">Both models are well-suited for datasets where features are well-structured and the target classes are separable (as in the breast cancer dataset).</font>

### Worst Performer:Decision Tree Classifier
-  <font size="5">Performed the worst among the five models with 93.86% accuracy.
Decision Trees can overfit without proper regularization, which likely reduced its generalization ability on the test set.</font>
-  <font size="5">Decision Trees are prone to overfitting, especially with datasets that have many features.  hyperparameter tuning might improve its performance.</font>
